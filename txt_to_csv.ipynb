```python
f = open('C:/Users/class/OneDrive/바탕 화면/processed/processed_korquad.txt', 'r', encoding = 'UTF8')
```

```python
dt = f.read().splitlines()
```

```python
type(dt)
```

```python
qas = []
paragraphs = []
for word in dt : 
    if '?' in word:
        qas.append(word)
    else: 
        paragraphs.append(word)
```

```python
print(qas)
```

```python
q = [] 
a = []
for i in qas: 
    aa=i.split('?')[0]
    q.append(aa)
    bb= i.split('?')[1]
    a.append(bb)
    
```

```python
print(a)
```

```python
import pandas as pd   ## csv 저장 
data= {'q': q, 'a':a}
df = pd.DataFrame(data)
df.to_csv('./wiki_qna.csv', encoding = 'utf-8-sig')
```

```python
para = {'paragraph': paragraphs}
df2 = pd.DataFrame(para)
df2.to_csv('./wiki_para.csv', encoding = 'utf-8-sig')
```
